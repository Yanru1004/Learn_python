```python
def show_result(func):
    def wrapper(data):
        print(f'Before: {data}')
        func(data)
        print(f'After: {data}')
    return wrapper
```

```python
#氣泡排序法

def bubble_sort(data):
    print(f'Before:{data}')
    for i in range(len(data)-1):
        for j in range(0,len(data)-1-i):
            
            if data[j] > data[j+1]:
                data[j],data[j+1] = data[j+1],data[j]
            #print(data)
    print(f'After: {data}')

bubble_sort([9,10,3,7,4,6,5,2,1,8])
```

```python
#插入排序法

def insertion_sort(data):
    print(f'Before:{data}')
    for i in range(1,len(data)):
        n = i
        while n > 0:
            if data[n] > data[n-1]:
                break
            else:
                data[n],data[n-1] = data[n-1],data[n]
            n -= 1
    print(f'After:{data}')

insertion_sort([9,10,3,7,4,6,5,2,1,8])
```

```python
#選擇排序法

def selection_sort(data):
    n = len(data)
    print(f'Before:{data}')
    #第幾次掃描迴圈
    for i in range(n-1):
        temp = data[i]
        min = data[i]
        min_index = i
        #掃描迴圈
        for j in range(i+1,n):
            #找出最小的數字及索引
            if data[j] < min:
                min = data[j]
                min_index = j
        #最小值與目前掃描起點索引交換
        data[i] = min
        data[min_index] = temp

    print(f'After:{data}')

selection_sort([9,10,3,7,4,6,5,2,1,8])
```

```python
#謝爾排序法

@show_result
def shell_sort(data):
    
    n = len(data)
    gap = n//2
    while gap > 0:
        for i in range(gap,n):
            temp = data[i]
            j = i
            while j >= gap and data[j-gap] > temp:
                data[j] = data[j-gap]
                j -= gap
            data[j] = temp
        gap //= 2
    return data


li = [9,10,3,7,4,6,5,2,1,8]

shell_sort(li)
```

```python
#合併排序法

def merge_sort(data):
    #分割    
    if len(data) <= 1:
        return data
    
    cut = len(data)//2
    left  = data[:cut]
    right = data[cut:]

    merge_sort(left)
    merge_sort(right)

    #合併
    left_i,right_i,data_i = 0,0,0

    while left_i < len(left) and right_i < len(right):
        #如果左邊較小
        if left[left_i] <= right[right_i]:
            data[data_i] = left[left_i]
            left_i += 1
            data_i += 1
        #如果右邊較小    
        else:
            data[data_i] = right[right_i]
            right_i += 1
            data_i += 1
        
    while left_i < len(left):
        data[data_i] = left[left_i]
        left_i += 1
        data_i += 1
    
    while right_i < len(right):
        data[data_i] = right[right_i]
        right_i += 1
        data_i += 1


data = [1,3,5,7,-2,4,6,8]
merge_sort(data)
print(data)

    
```
