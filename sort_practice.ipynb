```python
#氣泡排序法

def bubble_sort(li):
    for i in range(len(li),0,-1):
        for j in range(0,i-1):
            if li[j] > li[j+1]:
                #交換位置
                li[j],li[j+1] = li[j+1],li[j]
                print(f'{li}  {li[j]:>3} <---> {li[j+1]:>3}')

            else:
                print(f'{li}  {li[j]:>3} ----- {li[j+1]:>3}')

bubble_sort([8,2,6,10,4])
```

```python
#插入排序法

def insertion(li):
    for i in range(len(li)):
        n = i
        print(f'\n第{n+1}位 {li[n]} 開始 {li}')
        while (n!=0):
            if li[n] < li[n-1]:
                print(f'我是在第{n+1}位的 {li[n]} 前面是 {li[n-1]}，前面比我大，交換')
                print(f'\t{li} --->',end=' ')
                li[n],li[n-1] = li[n-1],li[n]
                print(f'{li}')
                n -= 1
            else:
                print(f'我是{li[n]} 前面是{li[n-1]}，前面比我小了，下一位')
                break
    print(li)


insertion([4,8,10,2,6])
```

```python
#選擇排序法

def selection_sort(li):
    for i in range(len(li)):
        _max = max(li[i:])
        _max_index = li.index(_max)
        print(f'串列{li} 最大的是:{_max} 在第{_max_index}位')

        if i != _max_index:
            li[i],li[_max_index] = li[_max_index],li[i]
            print(f'第{i}項 與 第{_max_index}項交換 串列{li}')
        
    print(f'{li}')

selection_sort([4,8,10,2,6])
```

```python
#選擇排序法

def selection_sort(li):
    for i in range(len(li)-1):
        temp = li[i]
        min = li[i]
        min_index = i
        for j in range(i+1,len(li)):
            if li[j] < min:
                min = li[j]
                min_index = j
        if min_index != i:
            li[i] = min
            li[min_index] = temp
        else:
            pass
    print(li)

selection_sort([4,9,10,2,6])
```

```python
#謝爾排序法
li = [5,7,6,1,8,4,3,2] 

def shell_sort(li):
    n = len(li)
    gap = n // 2

    while gap > 0:
        print(f'gap = {gap}開始')
        print(f'開始{gap} 到 {n-1} 迴圈')
        for i in range(gap,n):
            print(f'i = {i} j={i} temp設為{li[i]}')
            temp = li[i]
            j = i

            while j >= gap and li[j-gap] > temp:
                print(f'{li} 注意到索引{j-gap}的{li[j-gap]} 比 {temp}大')
                print(f"將索引{j} 取代為{li[j-gap]}",end=" ")
                li[j] = li[j-gap]
                print(f' 變成 {li}')
                j -= gap
            
            li[j] = temp
        gap //= 2
    
    print(li)

shell_sort(li)

        
```
