```python
def print_Matrix(Mat):
    for sub_mat in Mat:
        for e in sub_mat:
            print(f'{e: 4d}',end=' ')
        print("")


def show_Matrix(func):
    def wraper(*args):
        if len(args) == 2:
            print('Matrix A:')
            print_Matrix(args[0])
            print('Matrix B:')
            print_Matrix(args[1])
            print('Matrix C:')
            return print_Matrix(func(*args))
    return wraper

Matrix = list[list[int]]
```

```python
@show_Matrix
def sum_Matrix(Mat_A: Matrix,Mat_B: Matrix)->Matrix:
    m = len(Mat_A)
    n = len(Mat_A[0])

    Mat_C = [[None] * n for rows in range(m)]

    for i in range(m):
        for j in range(n):
            Mat_C[i][j] = Mat_A[i][j] + Mat_B[i][j]
    
    return Mat_C

A = [[1,3,5],[7,9,11],[13,15,17]]
B = [[9,8,7],[6,5,4] ,[3,2,1]]

sum_Matrix(A,B)
```

```python
@show_Matrix
def matrixMultiply(Mat_A:Matrix,Mat_B:Matrix):
    m = len(Mat_A)
    n = len(Mat_A[0])
    p = len(Mat_B[0])

    Mat_C = [[None] * m for p in range(p)]

    for i in range(m):
        for j in range(p):
            temp = 0
            for k in range(n):
                temp += Mat_A[i][k] * Mat_B[k][j]
            Mat_C[i][j] = temp
    return Mat_C

A = [[6,3,5],[8,9,7]]
B = [[5,10],[14,7],[6,8]]

matrixMultiply(A,B)
```
